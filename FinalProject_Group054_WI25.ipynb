# Abstract

This study aimed to investigate the relationship between social media usage and mental health, focusing particularly on students from the ages 17 -23. Given the increasing prevalence of social media in daily life, understanding its impact on mental well-being can be beneficial to us and our peers, considering how we ourselves are within the research scope. We analyzed survey data from students, examining correlations between the amount of time spent on social media platforms and indication of mental health wellness through a scoring system that we generated to reflect anxiety or depression. In addition, we explored demographic factors of gender and relationship status to determine if they play a part in social media usage and how such usage reflects upon their psychological well being.

The analysis suggested a weak correlation between social media usage and reported mental health issues. While there may be some connection, it is not significant enough to draw a definitive conclusion. However our study raises challenges such as how our data are voluntary surveys, allowing difficulty of accurate measuring in social media use. Additionally, other environmental, social, and personal factors that influence mental health were not fully accounted for in the datasets, which could have further impacted the results.

Overall, the findings indicate that social media usage may have some effect on mental health but it is likely just one of many contributing factors. The study suggests that future research should consider a broader range of influences and account for the confounding factors to better investigate this topic.

Gender has some inconsistencies, we will write a function to standardize the gender values to male, female, or nonbinary.

```python
# Standardize gender values
def standardize_gender(string):
    
    string = string.lower()
    string = string.strip()
    
    if 'nonbinary' in string:
        output = 'nonbinary'
    elif 'non-binary' in string:
        output = 'nonbinary'
    elif 'non binary' in string:
        output = 'nonbinary'
    elif 'unsure' in string:
        output = 'nonbinary'
    # Otherwise, if uncaught - keep as is
    else:
        output = string
    
    return output

df['Gender']= df['Gender'].apply(standardize_gender)
df['Gender'].unique()
```

Now let's use the data in the columns to calculate scores UCLA Loneliness Scale-8, General Anxiety Disorder 7 (GAD-7) Anxiety Severity, Patient Health Questionnaire-9 (PHQ-9) Score, Social Media Usage Scale, and Pittsburgh Sleep Quality Index (PSQI). These scales will be used later in our analysis and modeling.

The UCLA Loneliness Scale is calculated by first assigning scores from 1-4 for 6 columns and 4-1 for 2 other columns. These 8 columns are related to feelings of loneliness. Then, we will take the average of these scores to get the raw scale score. The UCLA-8 Scale is calculated using this equation: UCLA-8 (0-100) = (raw scale score – 1) * (100/3). The scores range from 0 to 100.

```python
# UCLA Loneliness Scale (UCLA-8), https://labs.dgsom.ucla.edu/hays/files/view/docs/surveys/ULS-8.pdf
def assign_score(frequency):
    if frequency == 'Never':
        return 1
    if frequency == 'Rarely':
        return 2
    if frequency == 'Sometimes':
        return 3
    if frequency == 'Often':
        return 4

def assign_alt_score(frequency):
    if frequency == 'Never':
        return 4
    if frequency == 'Rarely':
        return 3
    if frequency == 'Sometimes':
        return 2
    if frequency == 'Often':
        return 1
    
# Assign scores 1-4 for these columns
# Never = 1
# Rarely = 2
# Sometimes = 3
# Often = 4
columns_score = [
    'Lack_Companionship', 'No_One_To_Turn', 'Feeling_Left_Out', 
    'Feeling_Isolated', 'Unhappy_Withdrawn', 'People_Around_Not_With'
]

# Assign scores 4-1 for these columns
# Never = 4
# Rarely = 3
# Sometimes = 2
# Often = 1
columns_alt_score = ['Outgoing_Person', 'Can_Find_Companionship']

df2[columns_score] = df2[columns_score].applymap(assign_score)
df2[columns_alt_score] = df2[columns_alt_score].applymap(assign_alt_score)

# Average these scores to get the raw scale score
# UCLA-8 (0-100) = (raw scale score – 1) * (100/3)

df2['UCLA-8'] = (
    (((df2['Lack_Companionship'] + df2['No_One_To_Turn'] + df2['Feeling_Left_Out'] + df2['Feeling_Isolated'] + df2['Unhappy_Withdrawn'] + 
    df2['People_Around_Not_With'] + df2['Outgoing_Person'] + df2['Can_Find_Companionship']) / 8) - 1) * (100/3)
    )
```

The GAD-7 Anxiety Severity Scale is calculated by assigning scores to 7 columns related to anxiety and taking the sum of those scores. The scores range from 0 to 21. Scores of 0-4 are minimal anxiety, 5-9 are mild anxiety, 10-14 are moderate anxiety, and 15-21 are severe anxiety.

```python
# General Anxiety Disorder 7 (GAD-7) Anxiety Severity, https://adaa.org/sites/default/files/GAD-7_Anxiety-updated_0.pdf
def assign_anxiety_scores(x):
    if x == 'Not at all':
        return 0
    if x == 'Several days':
        return 1
    if x == 'Half days':
        return 2
    if x == 'Nearly everyday':
        return 3
    
# Assign scores 0-3 for these columns
anxiety_columns = [
    'Feeling_Anxious', 'Uncontrollable_Worry', 'Excessive_Worry', 'Trouble_Relaxing', 
    'Restlessness', 'Easily_Irritated', 'Fearful_Feeling'
]

df2[anxiety_columns] = df2[anxiety_columns].applymap(assign_anxiety_scores)

# Sum up the scores
# 0–4: minimal anxiety
# 5–9: mild anxiety
# 10–14: moderate anxiety
# 15–21: severe anxiety
df2['GAD-7'] = (
    df2['Feeling_Anxious'] + df2['Uncontrollable_Worry'] + df2['Excessive_Worry'] + df2['Trouble_Relaxing'] + 
    df2['Restlessness'] + df2['Easily_Irritated'] +  df2['Fearful_Feeling']
)
```

The PHQ-9 Scale is used to measure levels of depression. It is calculated by assigning scores to 9 columns related to behaviors of depression and taking the sum of those scores. Scores range from 0 to 27. Scores 0-4 are minimal depression, 5-9 are mild depression, 10-14 are moderate depression, 15-19 are moderately severe depression, and 20-27 are severe depression.

```python
# Patient Health Questionnaire-9 (PHQ-9) Score, used to examine depression levels
# https://med.stanford.edu/fastlab/research/imapp/msrs/_jcr_content/main/accordion/accordion_content3/download_256324296/file.res/PHQ9%20id%20date%2008.03.pdf
def convert_phq9(response):
    if response == 'Not at all':
        return 0
    elif response == 'Several days':
        return 1
    elif response == 'Half of days':
        return 2
    elif response == 'Nearly everyday':
        return 3
    else:
        return None  

phq9_columns = ['Low_Interest', 'Feeling_Down', 'Sleep_Issues', 'Low_Energy',
                'Appetite_Issues', 'Low_Self_Worth', 'Poor_Concentration',
                'Slow_Or_Restless', 'Self_Harm_Thoughts']

df2[phq9_columns] = df2[phq9_columns].applymap(convert_phq9)

# Calculating the final PHQ-9 score by summing
# 0-4 = Minimal depression
# 5-9 = Mild depression
# 10-14 = Moderate depression
# 15-19 = Moderately severe depression
# 20-27 = Severe depression
df2['PHQ-9'] = (df2['Low_Interest'] + df2['Feeling_Down'] +  df2['Sleep_Issues'] + df2['Low_Energy'] +
                df2['Appetite_Issues'] + df2['Low_Self_Worth'] + df2['Poor_Concentration'] +
                df2['Slow_Or_Restless'] + df2['Self_Harm_Thoughts'])
```

The Social Media Usage Scale is calculated by assigning scores to 13 columns related to social media usage and taking the sum of those scores.

```python
# Social Media Usage Scale
# Assign scores for columns
mappings = {
    'Device_Used': {'Mobile Phone': 4, 'Tablet': 3, 'Laptop': 2, 'Desktop': 1},
    'SM_Usage_Duration': {'More than 10 years': 4, '5-10 years': 3, '2-5 years': 2, 'Less than 2-year': 1},
    'Post_Frequency': {'3-5 per day': 4, '1-2 per day': 3, 'More than 5 per day': 2, 'Less than 1 per day': 1},
    'Daily_SM_Time': {'More than 5 hours': 4, '3-5 hours': 3, '1-3 hours': 2, 'Less than 1 hour': 1},
    'SM_Usage_Time': {'Frequently at anytime': 4, 'Night to late night': 2, 'Evening': 1, 'Afternoon': 1, 'Morning': 1},
    'Total_Friends': {'Less than 500': 1, '500-2000': 2, '2000-4000': 3, 'More than 4000': 4},
    'Known_Friends': {'Few of them': 4, 'Many of them': 3, 'Most of them': 2, 'All of them': 1},
    'Tagged_Groups': {'Less than 5': 1, '5-10': 2, '10-20': 3, 'More than 20': 4},
    'Peer_Pressure': {'Yes': 4, 'No': 0},
    'Emotional_Influence': {'Not at all': 0, 'Sometimes': 2, 'Always': 4},
    'Comparison': {'Never': 0, 'Most of the times': 2, 'All the times': 4},
    'Better_Without_SM': {'Yes': 4, 'No': 0},
    'Trying_To_Reduce': {'Not trying': 0, 'Trying to reduce the use': 2, 'Trying to stop the use': 4},
}

for col, mapping in mappings.items():
    df2[col] = df2[col].map(mapping)

df2['Social_Media_Usage_Scale'] = df2[list(mappings.keys())].sum(axis=1)
```

The PSQI Scale is a little bit more complicated. 7 columns related to sleeping behavior are assigned scores. These scores are divided into 4 different components and the components are summed to get the final PSQI Scale.

```python
# Pittsburgh Sleep Quality Index (PSQI)
# Define mappings for categorical columns
mappings2 = {
    'Sleep_Quality': {'Very good': 0, 'Fairly good': 1, 'Fairly bad': 2, 'Very bad': 3},
    'Sleep_Latency': {'Less than 15 minutes': 0, '15-30 minutes': 1, '31-60 minutes': 2, 'More than 60 minutes': 3},
    'Sleep_Onset_Issue': {'Not during last month': 0, 'Less than once a week': 1, 'Once or twice a week': 2, 'Three or more in week': 3},
    'Actual_Sleep_Hours': {'More than 8 hours': 0, '7 to 8 hours': 1, '4 to 6 hours': 2, 'Less than 4 hours': 3},
    'Sleep_Meds_Usage': {'Not during last month': 0, 'Less than once a week': 1, 'Once or twice a week': 2, 'Three or more in week': 3},
    'Missed_Sleep_Due_To_Events': {'Not during last month': 0, 'Less than once a week': 1, 'Once or twice a week': 2, 'Three or more in week': 3},
    'Event_Interference': {'Not during last month': 0, 'Less than once a week': 1, 'Once or twice a week': 2, 'Three or more in week': 3},
}

# Apply mappings
for col, mapping in mappings2.items():
    df2[col] = df2[col].map(mapping)

# Ensure numeric types for sleep efficiency calculation
df2['Actual_Sleep_Hours'] = pd.to_numeric(df2['Actual_Sleep_Hours'], errors='coerce')
df2['Time_In_Bed'] = pd.to_numeric(df2['Time_In_Bed'], errors='coerce')

# Component 2: Sum 'Sleep_Latency' and 'Sleep_Onset_Issue'
df2['Component_2'] = df2['Sleep_Latency'] + df2['Sleep_Onset_Issue']
df2['Component_2'] = np.select(
    [df2['Component_2'] == 0, df2['Component_2'].between(1, 2), df2['Component_2'].between(3, 4), df2['Component_2'].between(5, 6)],
    [0, 1, 2, 3]
)

# Component 4: Calculate Sleep Efficiency and categorize
df2['Sleep_Efficiency'] = df2['Actual_Sleep_Hours'] / df2['Time_In_Bed']
df2['Component_4'] = np.select(
    [df2['Sleep_Efficiency'] > 0.85, df2['Sleep_Efficiency'].between(0.75, 0.84), df2['Sleep_Efficiency'].between(0.65, 0.74), df2['Sleep_Efficiency'] < 0.65],
    [0, 1, 2, 3]
)

# Component 5: Sum disturbance scores
disturbance_cols2 = ['Midnight_Wakeups', 'Bathroom_Usage', 'Breathing_Issue', 'Loud_Snoring', 'Feeling_Cold', 'Feeling_Hot', 'Bad_Dreams', 'Pain_During_Sleep', 'Other_Sleep_Issues']
df2[disturbance_cols2] = df2[disturbance_cols2].replace({'Not during last month': 0, 'Less than once a week': 1, 'Once or twice a week': 2, 'Three or more in week': 3})
df2[disturbance_cols2] = df2[disturbance_cols2].apply(pd.to_numeric, errors='coerce')
df2['Disturbance_Sum'] = df2[disturbance_cols2].sum(axis=1)
df2['Component_5'] = np.select(
    [df2['Disturbance_Sum'] == 0, df2['Disturbance_Sum'].between(1, 9), df2['Disturbance_Sum'].between(10, 18), df2['Disturbance_Sum'].between(19, 27)],
    [0, 1, 2, 3]
)

# Component 7: Sum event-related disturbances
df2['Missed_Sleep_Sum'] = df2['Missed_Sleep_Due_To_Events'] + df2['Event_Interference']
df2['Component_7'] = np.select(
    [df2['Missed_Sleep_Sum'] == 0, df2['Missed_Sleep_Sum'].between(1, 2), df2['Missed_Sleep_Sum'].between(3, 4), df2['Missed_Sleep_Sum'].between(5, 6)],
    [0, 1, 2, 3]
)

# Sum all components for final PSQI score
df2['PSQI_Score'] = df2[['Sleep_Quality', 'Component_2', 'Actual_Sleep_Hours', 'Component_4', 'Component_5', 'Sleep_Meds_Usage', 'Component_7']].sum(axis=1)
```

## Inferential Modeling

We are going to use multiple linear regression using Ordinary Least Squares (OLS) Regression to measure the effect of social media usage and other predictors on loneliness, depression, and anxiety. 

Renamed the columns since OLS model does not take in the column name if it contains a number.

```python
df2.rename(columns={'UCLA-8': 'UCLA_Score'}, inplace=True)
df2.rename(columns={'GAD-7': 'GAD_Score'}, inplace=True)
df2.rename(columns={'PHQ-9': 'PHQ_Score'}, inplace=True)
```

```python
# Correlation between UCLA Loneliness Score and Social Media Usage + loneliness variables
mod = smf.ols(formula='UCLA_Score ~ Social_Media_Usage_Scale + No_One_To_Turn + Feeling_Isolated + Unhappy_Withdrawn + People_Around_Not_With', data=df2)
res = mod.fit()
print(res.summary())
```

The R-Squared score of 0.831 explains 83.1% variance in UCLA_Score, which indicates a strong fit. Social Media Usage has a low positive effect on UCLA_Score, which indicates that higher social media usage may lead to higher loneliness scores. However, it is not statistically significant as the p-value of 0.135 is greater than 0.05. The other variables all have a strong positive effect and are all statistically significant with p-values less than 0.05. Loneliness is most strongly predicted by feeling isolated, having no one to turn to, being unhappy/withdrawn, and feeling disconnected from people around you. Social media usage is not a significant predictor of loneliness in this model.

```python
# Correlation between Anxiety Scores and Social Media Usage + anxiety variables
mod = smf.ols(formula='GAD_Score ~ Social_Media_Usage_Scale + Feeling_Anxious + Excessive_Worry + Trouble_Relaxing + Fearful_Feeling', data=df2)
res = mod.fit()
print(res.summary())
```

The R-Squared score of 0.957 indicates a very strong fit. Social Media Usage has a very small positive effect on anxiety scores and is not statistically significant which suggests social media usage does not meaningfully predict anxiety. It also is not statistically significant. Anxiety is most strongly predicted by feeling anxious, fearful, having excessive worry, and trouble relaxing. These variables all are statistically significant. Social media usage is not a significant predictor of anxiety.

```python
# Correlation between Depression Scores and Social Media Usage + depression behavior variables
mod = smf.ols(formula='PHQ_Score ~ Social_Media_Usage_Scale + Low_Interest + Feeling_Down + Sleep_Issues + Appetite_Issues + Poor_Concentration + Self_Harm_Thoughts', data=df2)
res = mod.fit()
print(res.summary())
```

The R-Squared score of 0.938 indicates a very strong fit. Social Media Usage has a very small effect on depression scores in this model. Depression severity (PHQ_Score) is most strongly influenced by appetite issues, sleep problems, feeling down, poor concentration, and self-harm thoughts. All these variables are statistically significant. Social media usage has no meaningful effect on depression in this model.

Based on these models, social media usage does not appear to have a significant effect loneliness, anxiety, and depression scores. This suggests that other factors are much more important in determining mental health outcomes than social media use alone.

# Ethics & Privacy

Privacy and Data Protection

- This dataset from Kaggle 'Social Media and Mental Health' was collected through a voluntary survey, meaning that participants consented in sharing certain information.
- This dataset also does not include personal identifiers such as names, emails, or ids. However, it does contain age, gender, relationship status, and occupation status.
- While there are no personal identifiers in this dataset their is a possibility of bias as a voluntary survey introduces self-selection bias.
Since participants self-reported their social media usage and mental health, this could lead the data to lean more towards people who are more open in sharing experiences.
- We will set out to detect these specific biases by analyzing the demographic distribution in the dataset by identifying underrepresented or over represented groups, such as age or gender imbalances. During the analysis we plan to search for sampling biases by reviewing response rates across different groups in order to make sure that no single group is above the rest. After analyzing the data we will review our conclusions if we find potential biases we will state them at the end.
- In the data, it seems that students make up a large portion of the demographic, which could be a problem in representing the smaller portion of the demographic.
- We will handle issues mentioned by instead focusing only on University students and School students, as these two occuppation statuses make up a large portion of the dataset. This will help make sure our findings are more relevant and applicable to the majority of the participants.
- We will analzye the response rates across different demographic groups to look for any imbalances. If certain groups are underrepresented we will apply Statistical weighting to adjust for the imbalances. This way the final analysis will represent everyone.
- Mental Health data may be subject to both underreporting and overreporting, this can affect interpretation as it can make certain groups experiences disproportionately larger than the others which may influence the findings.
- Excluding non-students odes not introduce additional bias because our research question focuses only on students mental health. The fact that the datasets largely contains students makes it so that our findings will not be skewed by other groups which we will drop from the datasets.

# Discussion and Conclusion

Our study aimed to explore the relationship between social media usage and mental health, particularly in students. Our findings suggested a weak correlation between social media usage and reported mental health impacts. Similarly, the analysis of dataset 2 revealed only a minor correlation, preventing us from drawing a definitive conclusion. As we looked into the demographic factors of gender and relationship status, the correlation of social media usage and these points, along with how these participants scored on the metric of mental health, only proposes a weak relation and is not significant to suggest that these platforms impact one's psychology. There are many contributing factors to mental health and social media is most likely only one of the many factors, however, it also suggests that other environmental, social, and personal influences can play a significant role in mental health that the datasets cannot take into account.

One of the biggest challenges we faced was dataset availability. Public datasets that specifically address the correlation between social media usage and mental health are hard to find. Those that do exist often lack depth or certain factors to answer our research question. As a result, we had to rely on voluntary surveys. One problem that arises with this format of surveys is that they only contain data that are self reported. Self-reported data can be influenced by personal biases, resulting in biased results. They also often only contain data of strongly opinionated individual’s, making the results less generalized and applicable to the society. In addition, the datasets do not account for individuals who already experience mental health challenges, creating an underrepresentation in our research as there is no sure way to know surveyees’ mental health status. 


Measuring social media in itself was another challenge as the two datasets inlcuded mixed response formats and ranges. At times the format would use integers and other times it would use ranges in the form of a string. This made it harder to analyze the data as we had to create our own unified scale for the exploratory data analysis. Without this unified scale our exploratory data anlysis would have consisted of graphs with x and y scales from ranges 1 to 5. This would have provided little to nothing for our research question as we could not highlight or visualize any trends. Additionally, mental health is infuenced by many environmental and personal factors that go beyond social media. Since our study and data only focused on social media, we were not able to take into account these external factors. 

Despite these limitations our research contributes the ongoing discussion about social media's role in mental health. While our findings were weak we beloive that with a larger data set and a structured data collection method that we could find evidence that social medai plays a role in the mental health of students.
